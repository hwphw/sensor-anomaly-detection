```python
import pandas as pd 

df = pd.read_csv('../data/train_FD001.txt', sep='\s+', header=None)
df.shape
df.head()
```

```python
columns = [
    'unit', 'time',
    'op_setting_1', 'op_setting_2', 'op_setting_3',
    'sensor_1', 'sensor_2', 'sensor_3', 'sensor_4', 'sensor_5',
    'sensor_6', 'sensor_7', 'sensor_8', 'sensor_9', 'sensor_10',
    'sensor_11', 'sensor_12', 'sensor_13', 'sensor_14', 'sensor_15',
    'sensor_16', 'sensor_17', 'sensor_18', 'sensor_19', 'sensor_20', 'sensor_21'
]

df.columns = columns 
df.head()
```

```python
import matplotlib.pyplot as plt 

# 하나의 센서 값 그래프 그리기 (시각화)
unit1 = df[df['unit'] == 1] 

plt.figure(figsize=(12,4))
plt.plot(unit1['time'], unit1['sensor_2'])
plt.title('Unit 1 - Sensor 2 Trend')
plt.xlabel('Time')
plt.ylabel('Sensor Value')
plt.show()
```

```python
sensors = ['sensor_1', 'sensor_2', 'sensor_3', 'sensor_4', 'sensor_5']

plt.figure(figsize=(12,8))

for s in sensors: 
    plt.plot(unit1['time'], unit1[s], label=s)

plt.legend()
plt.title("Unit 1 - Sensor Patterns")
plt.show()
```

```python
import matplotlib.pyplot as plt 
# 전체 센서 자동 시각화 함수 
def plot_unit_sensors(df, unit_id):
    sensors = [col for col in df.columns if col.startswith('sensor_')]
    unit_df = df[df['unit'] == unit_id]

    plt.figure(figsize=(15, 30))

    for i, s in enumerate(sensors, 1):
        plt.subplot(7, 3, i)    # 21개의 센서 -> 7행 3열로 배치 
        plt.plot(unit_df['time'], unit_df[s])
        plt.title(s)
        plt.tight_layout()

    plt.show()

plot_unit_sensors(df, 1)
```
